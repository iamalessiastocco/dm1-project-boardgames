# Analisi Outliers e Correlazioni

In questa sezione analizziamo la presenza di outliers e le correlazioni tra attributi per decidere se mantenere o rimuovere alcune variabili.

```python
# BOXPLOT CONDIZIONALI - Vedere outliers per ogni categoria di Rating
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Boxplot Condizionali per Rating (Identificazione Outliers)', fontsize=16)

# Lista di attributi numerici da analizzare
attributi = ['ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime', 'ComMaxPlaytime', 'YearPublished']

for idx, attr in enumerate(attributi):
    row = idx // 3
    col = idx % 3
    sns.boxplot(data=df, x='Rating', y=attr, order=['Low', 'Medium', 'High'], 
                palette='Set2', ax=axes[row, col])
    axes[row, col].set_title(f'{attr} per Rating')
    axes[row, col].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ISTOGRAMMI DETTAGLIATI - Distribuzione di ogni attributo numerico
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
fig.suptitle('Distribuzioni degli Attributi Numerici (Identificazione Outliers)', fontsize=16)

attributi_numerici = ['ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime', 
                      'ComMaxPlaytime', 'YearPublished', 'MinPlayers', 'MaxPlayers', 'ComMinAge']

for idx, attr in enumerate(attributi_numerici):
    row = idx // 3
    col = idx % 3
    
    # Istogramma
    axes[row, col].hist(df[attr].dropna(), bins=50, color='steelblue', edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'Distribuzione {attr}')
    axes[row, col].set_xlabel(attr)
    axes[row, col].set_ylabel('Frequenza')
    
    # Aggiungiamo linee verticali per media e mediana
    mean_val = df[attr].mean()
    median_val = df[attr].median()
    axes[row, col].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Media: {mean_val:.2f}')
    axes[row, col].axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Mediana: {median_val:.2f}')
    axes[row, col].legend()

plt.tight_layout()
plt.show()
```

```python
# SCATTERPLOT 1: Attributi RANK vs CAT
# Cerchiamo colonne che contengono 'rank' e 'cat' nel nome
rank_cols = [col for col in df.columns if 'rank' in col.lower()]
cat_cols = [col for col in df.columns if 'cat' in col.lower()]

print("Attributi di Rank trovati:", rank_cols)
print("Attributi di Cat trovati:", cat_cols)

# Combiniamo rank e cat insieme
rank_cat_cols = rank_cols + cat_cols

if len(rank_cat_cols) >= 2:
   
    # Matrice di correlazione per rank e cat
    rank_cat_corr = df[rank_cat_cols].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(rank_cat_corr, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0, square=True)
    plt.title('Correlazione tra Attributi RANK e CAT')
    plt.tight_layout()
    plt.show()
    
    # Stampiamo correlazioni alte
    print("\n=== CORRELAZIONI ALTE tra RANK e CAT (|r| > 0.7) ===")
    for i in range(len(rank_cat_corr.columns)):
        for j in range(i+1, len(rank_cat_corr.columns)):
            corr_val = rank_cat_corr.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"{rank_cat_corr.columns[i]} <-> {rank_cat_corr.columns[j]}: {corr_val:.3f}")
else:
    print("Non abbastanza attributi trovati")
```

```python
# SCATTERPLOT 3: Attributi di Popolarità/Engagement
# NumOwned, NumWant, NumWish, NumUserRatings, NumWeightVotes
popularity_cols = ['NumOwned', 'NumWant', 'NumWish', 'NumUserRatings', 'NumWeightVotes']

# Verifichiamo quali colonne esistono effettivamente
existing_popularity = [col for col in popularity_cols if col in df.columns]
print("Attributi di Popolarità trovati:", existing_popularity)

if len(existing_popularity) >= 2:
    # Creiamo pairplot
    fig = sns.pairplot(df[existing_popularity], diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20})
    fig.fig.suptitle('Scatterplot Matrix - Attributi di Popolarità/Engagement', y=1.01, fontsize=14)
    plt.tight_layout()
    plt.show()
    
    # Matrice di correlazione
    pop_corr = df[existing_popularity].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(pop_corr, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0, square=True)
    plt.title('Correlazione tra Attributi di Popolarità')
    plt.tight_layout()
    plt.show()
    
    # Stampiamo correlazioni alte
    print("\n=== CORRELAZIONI ALTE tra attributi di Popolarità (|r| > 0.7) ===")
    for i in range(len(pop_corr.columns)):
        for j in range(i+1, len(pop_corr.columns)):
            corr_val = pop_corr.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"{pop_corr.columns[i]} <-> {pop_corr.columns[j]}: {corr_val:.3f}")
else:
    print("Non abbastanza attributi di popolarità trovati")
```

```python
# SCATTERPLOT 4: Attributi di Tempo di Gioco
# ComMinPlaytime, ComMaxPlaytime, MfgAgeRec
time_cols = ['ComMinPlaytime', 'ComMaxPlaytime', 'MfgAgeRec']

# Verifichiamo quali colonne esistono effettivamente
existing_time = [col for col in time_cols if col in df.columns]
print("Attributi di Tempo trovati:", existing_time)

if len(existing_time) >= 2:
    # Creiamo pairplot
    fig = sns.pairplot(df[existing_time], diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20})
    fig.fig.suptitle('Scatterplot Matrix - Attributi di Tempo/Età', y=1.01, fontsize=14)
    plt.tight_layout()
    plt.show()
    
    # Matrice di correlazione
    time_corr = df[existing_time].corr()
    plt.figure(figsize=(8, 6))
    sns.heatmap(time_corr, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0, square=True)
    plt.title('Correlazione tra Attributi di Tempo/Età')
    plt.tight_layout()
    plt.show()
    
    # Stampiamo correlazioni alte
    print("\n=== CORRELAZIONI ALTE tra attributi di Tempo (|r| > 0.7) ===")
    for i in range(len(time_corr.columns)):
        for j in range(i+1, len(time_corr.columns)):
            corr_val = time_corr.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"{time_corr.columns[i]} <-> {time_corr.columns[j]}: {corr_val:.3f}")
else:
    print("Non abbastanza attributi di tempo trovati")
```

```python
# BOXPLOT OUTLIERS - Identificazione statistica degli outliers
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
fig.suptitle('Boxplot per Identificazione Outliers', fontsize=16)

for idx, attr in enumerate(attributi_numerici):
    row = idx // 3
    col = idx % 3
    
    # Boxplot
    bp = axes[row, col].boxplot(df[attr].dropna(), vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    
    axes[row, col].set_title(f'{attr}')
    axes[row, col].set_ylabel('Valore')
    axes[row, col].grid(True, alpha=0.3)
    
    # Calcolo statistiche IQR per identificare outliers
    Q1 = df[attr].quantile(0.25)
    Q3 = df[attr].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[attr] < lower_bound) | (df[attr] > upper_bound)][attr]
    
    # Aggiungiamo testo con info
    axes[row, col].text(0.5, 0.95, f'Outliers: {len(outliers)}', 
                        transform=axes[row, col].transAxes, 
                        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

# Stampiamo un report dettagliato degli outliers
print("\n=== REPORT OUTLIERS (metodo IQR: Q1-1.5*IQR, Q3+1.5*IQR) ===\n")
for attr in attributi_numerici:
    Q1 = df[attr].quantile(0.25)
    Q3 = df[attr].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[attr] < lower_bound) | (df[attr] > upper_bound)][attr]
    
    print(f"{attr}:")
    print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"  Range normale: [{lower_bound:.2f}, {upper_bound:.2f}]")
    print(f"  Numero outliers: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")
    if len(outliers) > 0:
        print(f"  Min outlier: {outliers.min():.2f}, Max outlier: {outliers.max():.2f}")
    print()
```

```python
# VIOLINPLOT - Distribuzione + densità + outliers
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Violinplot - Distribuzione e Densità per Rating', fontsize=16)

attributi_violin = ['ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime', 'ComMaxPlaytime', 'YearPublished']

for idx, attr in enumerate(attributi_violin):
    row = idx // 3
    col = idx % 3
    
    sns.violinplot(data=df, x='Rating', y=attr, order=['Low', 'Medium', 'High'], 
                   palette='muted', ax=axes[row, col])
    axes[row, col].set_title(f'{attr} - Distribuzione per Rating')
    axes[row, col].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
