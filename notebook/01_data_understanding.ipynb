```python
# Selezioniamo solo le colonne numeriche principali per non fare confusione
cols_analysis = ['Rating', 'ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime', 'YearPublished']

# Nota: Rating è testo (Low/Med/High), dobbiamo trasformarlo al volo in numeri SOLO per questo grafico
# Creiamo una copia temporanea per non toccare i dati originali
df_temp = df.copy()
df_temp['Rating_Num'] = df_temp['Rating'].map({'Low':0, 'Medium':1, 'High':2})

# Calcolo correlazione
correlation = df_temp[['Rating_Num', 'ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime']].corr()

# Visualizzazione (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice di Correlazione")
plt.show()
```

```python
# 1. Istogramma ANNO (Per vedere l'errore -3500)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Usiamo bins=50 per avere più dettaglio
plt.hist(df['YearPublished'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribuzione Anno di Pubblicazione")
plt.xlabel("Anno")
plt.ylabel("Quanti giochi")

# 2. Istogramma POPOLARITÀ (NumOwned)
plt.subplot(1, 2, 2)
plt.hist(df['NumOwned'], bins=50, color='orange', edgecolor='black')
plt.title("Distribuzione Popolarità (Chi possiede il gioco?)")
plt.xlabel("Numero possessori")
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Boxplot GIOCATORI
plt.subplot(1, 2, 1)
sns.boxplot(x=df['MaxPlayers'], color='lightgreen')
plt.title("Boxplot MaxPlayers (Cerca il 999!)")

# Boxplot DURATA
plt.subplot(1, 2, 2)
sns.boxplot(x=df['ComMaxPlaytime'], color='salmon')
plt.title("Boxplot Durata (Cerca i tempi infiniti)")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Usiamo scatterplot di Seaborn che è più facile di matplotlib puro
sns.scatterplot(data=df, 
                x='ComWeight', 
                y='NumOwned', 
                hue='Rating', # Colora i punti in base al voto
                hue_order=['Low', 'Medium', 'High'], # Ordine colori
                alpha=0.6) # Trasparenza

plt.title("Difficoltà vs Popolarità (Colorato per Voto)")
plt.xlabel("Complessità (1=Easy, 5=Hard)")
plt.ylabel("Quante persone lo hanno (Popolarità)")
# Limitiamo l'asse Y perché altrimenti i giochi famosissimi schiacciano tutto
plt.ylim(0, 20000) 
plt.show()
```
