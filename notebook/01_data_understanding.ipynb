# Analisi Outliers e Correlazioni

In questa sezione analizziamo la presenza di outliers e le correlazioni tra attributi per decidere se mantenere o rimuovere alcune variabili.

```python
# SCATTERPLOT 1: Attributi RANK vs CAT
# Cerchiamo colonne che contengono 'rank' e 'cat' nel nome
rank_cols = [col for col in df.columns if 'rank' in col.lower()]
cat_cols = [col for col in df.columns if 'cat' in col.lower()]

print("Attributi di Rank trovati:", rank_cols)
print("Attributi di Cat trovati:", cat_cols)

# Combiniamo rank e cat insieme
rank_cat_cols = rank_cols + cat_cols

if len(rank_cat_cols) >= 2:
   
    # Matrice di correlazione per rank e cat
    rank_cat_corr = df[rank_cat_cols].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(rank_cat_corr, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0, square=True)
    plt.title('Correlazione tra Attributi RANK e CAT')
    plt.tight_layout()
    plt.show()
    
    # Stampiamo correlazioni alte
    print("\n=== CORRELAZIONI ALTE tra RANK e CAT (|r| > 0.7) ===")
    for i in range(len(rank_cat_corr.columns)):
        for j in range(i+1, len(rank_cat_corr.columns)):
            corr_val = rank_cat_corr.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"{rank_cat_corr.columns[i]} <-> {rank_cat_corr.columns[j]}: {corr_val:.3f}")
else:
    print("Non abbastanza attributi trovati")
```

```python
# SCATTERPLOT 3: Attributi di Popolarità/Engagement
# NumOwned, NumWant, NumWish, NumUserRatings, NumWeightVotes
popularity_cols = ['NumOwned', 'NumWant', 'NumWish', 'NumUserRatings', 'NumWeightVotes']

# Verifichiamo quali colonne esistono effettivamente
existing_popularity = [col for col in popularity_cols if col in df.columns]
print("Attributi di Popolarità trovati:", existing_popularity)

if len(existing_popularity) >= 2:
    # Creiamo pairplot
    fig = sns.pairplot(df[existing_popularity], diag_kind='kde', plot_kws={'alpha': 0.6, 's': 20})
    fig.fig.suptitle('Scatterplot Matrix - Attributi di Popolarità/Engagement', y=1.01, fontsize=14)
    plt.tight_layout()
    plt.show()
    
    # Matrice di correlazione
    pop_corr = df[existing_popularity].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(pop_corr, annot=True, fmt='.2f', cmap='RdYlBu_r', center=0, square=True)
    plt.title('Correlazione tra Attributi di Popolarità')
    plt.tight_layout()
    plt.show()
    
    # Stampiamo correlazioni alte
    print("\n=== CORRELAZIONI ALTE tra attributi di Popolarità (|r| > 0.7) ===")
    for i in range(len(pop_corr.columns)):
        for j in range(i+1, len(pop_corr.columns)):
            corr_val = pop_corr.iloc[i, j]
            if abs(corr_val) > 0.7:
                print(f"{pop_corr.columns[i]} <-> {pop_corr.columns[j]}: {corr_val:.3f}")
else:
    print("Non abbastanza attributi di popolarità trovati")
```

```python
# Selezioniamo solo le colonne numeriche principali per non fare confusione VEDO SE TENERLO
cols_analysis = ['Rating', 'ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime', 'YearPublished']

# Nota: Rating è testo (Low/Med/High), dobbiamo trasformarlo al volo in numeri SOLO per questo grafico
# Creiamo una copia temporanea per non toccare i dati originali
df_temp = df.copy()
df_temp['Rating_Num'] = df_temp['Rating'].map({'Low':0, 'Medium':1, 'High':2})

# Calcolo correlazione
correlation = df_temp[['Rating_Num', 'ComWeight', 'NumOwned', 'NumWant', 'ComMinPlaytime']].corr()

# Visualizzazione (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice di Correlazione")
plt.show()
```

```python
# SCATTERPLOT SPECIFICI - Relazioni interessanti
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. ComMinPlaytime vs ComMaxPlaytime (dovrebbero essere molto correlati)
sns.scatterplot(data=df, x='ComMinPlaytime', y='ComMaxPlaytime', hue='Rating', 
                hue_order=['Low', 'Medium', 'High'], alpha=0.6, ax=axes[0, 0])
axes[0, 0].set_title('Tempo Minimo vs Massimo di Gioco')
axes[0, 0].set_xlabel('Tempo Minimo (min)')
axes[0, 0].set_ylabel('Tempo Massimo (min)')

# 2. NumOwned vs NumWant (popolarità attuale vs desiderio)
sns.scatterplot(data=df, x='NumOwned', y='NumWant', hue='Rating', 
                hue_order=['Low', 'Medium', 'High'], alpha=0.6, ax=axes[0, 1])
axes[0, 1].set_title('Possessori vs Desideri')
axes[0, 1].set_xlabel('Numero Possessori')
axes[0, 1].set_ylabel('Numero Desideranti')
axes[0, 1].set_xlim(0, 30000)
axes[0, 1].set_ylim(0, 15000)

# 3. ComWeight vs ComMinPlaytime (complessità vs tempo)
sns.scatterplot(data=df, x='ComWeight', y='ComMinPlaytime', hue='Rating', 
                hue_order=['Low', 'Medium', 'High'], alpha=0.6, ax=axes[1, 0])
axes[1, 0].set_title('Complessità vs Tempo Minimo')
axes[1, 0].set_xlabel('Complessità')
axes[1, 0].set_ylabel('Tempo Minimo (min)')

# 4. MaxPlayers vs NumOwned (numero giocatori vs popolarità)
sns.scatterplot(data=df, x='MaxPlayers', y='NumOwned', hue='Rating', 
                hue_order=['Low', 'Medium', 'High'], alpha=0.6, ax=axes[1, 1])
axes[1, 1].set_title('Max Giocatori vs Popolarità')
axes[1, 1].set_xlabel('Numero Massimo Giocatori')
axes[1, 1].set_ylabel('Numero Possessori')
axes[1, 1].set_xlim(0, 20)  # Escludiamo il valore 999
axes[1, 1].set_ylim(0, 30000)

plt.tight_layout()
plt.show()
```

## Analisi Outliers con Boxplot - Attributi Critici

Analizziamo gli attributi dove ci aspettiamo outliers o valori non realistici che potrebbero essere errori di data entry.

```python
# Prima verifichiamo quali colonne esistono nel dataset
print("Colonne disponibili nel dataset:")
print(df.columns.tolist())

# Cerchiamo le colonne che potrebbero corrispondere a AvgPlaytime e GameAge
# AvgPlaytime potrebbe essere la media tra ComMinPlaytime e ComMaxPlaytime
# GameAge potrebbe essere calcolato da YearPublished

# Verifichiamo se esiste AvgPlaytime o se dobbiamo crearla
if 'AvgPlaytime' in df.columns:
    print("\nTrovata colonna AvgPlaytime")
else:
    print("\nColonna AvgPlaytime non trovata - calcolerò la media tra ComMinPlaytime e ComMaxPlaytime")
    df['AvgPlaytime'] = (df['ComMinPlaytime'] + df['ComMaxPlaytime']) / 2

# Verifichiamo se esiste GameAge o se dobbiamo crearla
if 'GameAge' in df.columns:
    print("Trovata colonna GameAge")
else:
    print("Colonna GameAge non trovata - calcolerò l'età dal YearPublished")
    current_year = 2025  # Anno corrente
    df['GameAge'] = current_year - df['YearPublished']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Analisi Outliers con Scala Logaritmica (dove necessario)', fontsize=16)

# --- A. AvgPlaytime (SCALA LOGARITMICA) ---
sns.boxplot(data=df, y='AvgPlaytime', ax=axes[0,0], color='lightcoral')
axes[0,0].set_title('AvgPlaytime (Scala Logaritmica)', fontweight='bold')
axes[0,0].set_yscale('log') # <--- QUESTA RIGA RISOLVE IL PROBLEMA DELLO SCHIACCIAMENTO
axes[0,0].set_ylabel('Minuti (Log Scale)')
# Aggiungiamo etichette per gli outliers estremi
max_play = df['AvgPlaytime'].max()
axes[0,0].text(0, max_play, f' Max: {max_play:.0f} min', ha='center', va='bottom', color='red', fontweight='bold')

# --- B. GameAge ---
sns.boxplot(data=df, y='GameAge', ax=axes[0,1], color='lightblue')
axes[0,1].set_title('GameAge (Anni)', fontweight='bold')
# Qui non usiamo il log, ma limitiamo la vista se ci sono errori assurdi (tipo anni negativi)
axes[0,1].set_ylim(bottom=-5) 

# --- C. MinPlayers vs MaxPlayers (Side by Side) ---
# Usiamo un formato "melted" per metterli vicini nello stesso grafico
players_data = df[['MinPlayers', 'MaxPlayers']].melt(var_name='Type', value_name='Count')
sns.boxplot(data=players_data, x='Type', y='Count', ax=axes[1,0], palette='Set2')
axes[1,0].set_title('Confronto Giocatori Min vs Max', fontweight='bold')
axes[1,0].set_yscale('log') # Utile se hai giochi da 100+ giocatori

# --- D. Popolarità (NumOwned) ---
sns.boxplot(data=df, y='NumOwned', ax=axes[1,1], color='gold')
axes[1,1].set_title('NumOwned (Scala Logaritmica)', fontweight='bold')
axes[1,1].set_yscale('log') # Fondamentale per vedere la distribuzione dei "best seller"

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8)) #Errori logici, giochi che hanno dati impossibili

# Scatter plot
sns.scatterplot(data=df, x='MinPlayers', y='MaxPlayers', alpha=0.6, edgecolor=None)

# Linea diagonale (y = x)
# Tutto ciò che è SOTTO questa linea rossa è un errore (Min > Max)
max_val = max(df['MinPlayers'].max(), df['MaxPlayers'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', linewidth=2, label='Linea di Correttezza (Min=Max)')

plt.title('Controllo Coerenza: Min Players vs Max Players', fontsize=14)
plt.xlabel('Min Players')
plt.ylabel('Max Players')
plt.xscale('log') # Usiamo log anche qui per gestire gli estremi
plt.yscale('log')
plt.legend()

# Evidenziamo l'area di errore
plt.fill_between([0, max_val], [0, max_val], color='red', alpha=0.1, label='Area Errore (Min > Max)')

plt.grid(True, which="both", ls="-", alpha=0.2)
plt.show()

# Contiamo quanti errori ci sono
errori_logici = df[df['MinPlayers'] > df['MaxPlayers']].shape[0]
print(f"Numero di giochi con errore logico (Min > Max): {errori_logici}")
```
